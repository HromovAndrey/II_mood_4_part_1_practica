```python
import torch

def f(x):
    return x**2 / 3 + torch.sin(torch.pi * x)

x = torch.tensor([2.0], requires_grad=True)
learning_rate = 0.01

for _ in range(1000):
    y = f(x)
    y.backward()
    with torch.no_grad():
        x -= learning_rate * x.grad
        x.grad.zero_()

print("Мінімум функції досягається в точці x =", x.item())
```

```python
import torch

def f1(x, y):
    return 418.9829*2 - x*torch.sin(x.abs().sqrt()) - y*torch.sin(y.abs().sqrt())

def f2(x, y):
    return 2*x**2 - 1.05*x**4 + 1/6*x**6 + x*y + y**2

def f3(x, y):
    return 106.7645 + torch.sin(x)*torch.exp((1-torch.cos(y))**2) + \
           torch.cos(y)*torch.exp((1-torch.sin(x))**2) + \
           (x - y)**2

def gradient_descent(func, lr=0.01, steps=1000):
    x = torch.tensor([2.0], requires_grad=True)
    y = torch.tensor([2.0], requires_grad=True)
    
    for _ in range(steps):
        z = func(x, y)
        z.backward()
        with torch.no_grad():
            x -= lr * x.grad
            y -= lr * y.grad
            x.grad.zero_()
            y.grad.zero_()
    return x.item(), y.item()

x_min1, y_min1 = gradient_descent(f1)
x_min2, y_min2 = gradient_descent(f2)
x_min3, y_min3 = gradient_descent(f3)

print("Мінімум f1 досягається в точці (x, y) = ({}, {})".format(x_min1, y_min1))
print("Мінімум f2 досягається в точці (x, y) = ({}, {})".format(x_min2, y_min2))
print("Мінімум f3 досягається в точці (x, y) = ({}, {})".format(x_min3, y_min3))
```
